```python
#####TASK 1
import numpy as np
import sklearn.metrics
import matplotlib.pyplot as plt
from sklearn.metrics import normalized_mutual_info_score
from sklearn.cluster import KMeans
from keras.datasets import fashion_mnist

#Loading the dataset
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

#Data normalization
max_value = float(x_train.max())
x_train = x_train.astype('float32') / max_value
x_test = x_test.astype('float32') / max_value

#Data preprocessing
x_train = x_train.reshape((len(x_train), np.prod(x_train.shape[1:])))
x_test = x_test.reshape((len(x_test), np.prod(x_test.shape[1:])))

#Determining clusters using Elbow method
clusters = []

inertia = []
for n in range(4,20):
    kmeans = KMeans(n_clusters = n, n_init = 20, n_jobs = 4,random_state = 1)
    kmeans.fit(x_train)
    clusters.append(n)
    inertia.append(kmeans.inertia_)

#Data visualization
plt.plot(clusters,inertia) 
plt.title('Inertia vs Clusters') 
plt.ylabel('Inertia') 
plt.xlabel('Clusters') 
plt.show()

#Cluster evaluation
kmeans = KMeans(n_clusters = 10, n_init = 20, n_jobs =4)
kmeans.fit(x_train)
pred = kmeans.predict(x_test)
acc = normalized_mutual_info_score(y_test, pred)
print('Accuracy is:',acc * 100)      
```

```python
#####TASK 2
import numpy as np
import keras
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, normalized_mutual_info_score
from sklearn.utils.linear_assignment_ import linear_assignment
from sklearn.cluster import KMeans
from keras.models import Model, Sequential
from keras.layers import Input, Dense, Conv2D, MaxPooling2D, UpSampling2D, Flatten, Reshape
from keras import regularizers
from keras.datasets import fashion_mnist

#Loading the dataset
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

#Data normalization
max_value = float(x_train.max())
x_train = x_train.astype('float32') / max_value
x_test = x_test.astype('float32') / max_value

#Data preprocessing
x_train = x_train.reshape((len(x_train), np.prod(x_train.shape[1:])))
x_test = x_test.reshape((len(x_test), np.prod(x_test.shape[1:])))

#Building autoencoder
input_dim = x_train.shape[1]
encoding_dim = 32

autoencoder = Sequential()

#Encoder layers
autoencoder.add(Dense(4 * encoding_dim, input_shape=(input_dim,), activation='relu'))
autoencoder.add(Dense(2 * encoding_dim, activation='relu'))
autoencoder.add(Dense(encoding_dim, activation='relu'))

#Decoder layers
autoencoder.add(Dense(2 * encoding_dim, activation='relu'))
autoencoder.add(Dense(4 * encoding_dim, activation='relu'))
autoencoder.add(Dense(input_dim, activation='sigmoid'))

#Encoder Model
input_img = Input(shape=(input_dim,))
encoder_layer1 = autoencoder.layers[0]
encoder_layer2 = autoencoder.layers[1]
encoder_layer3 = autoencoder.layers[2]
encoder = Model(input_img, encoder_layer3(encoder_layer2(encoder_layer1(input_img))))

#Model Compilation
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')
history_train = autoencoder.fit(x_train, x_train,
                epochs=15,
                batch_size=256,
                validation_data=(x_test, x_test))

#Data visualization
loss = history_train.history['loss']
val_loss = history_train.history['val_loss']
epochs = range(15)
plt.figure()
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.ylabel('Loss') 
plt.xlabel('Epochs') 
plt.legend()
plt.show()

#Auto Encoder with Kmeans clustering
pred_auto_train = encoder.predict(x_train)
pred_auto = encoder.predict(x_test)
km = KMeans(n_jobs=-1, n_clusters=10, n_init=20, random_state = 1)
km.fit(pred_auto_train)
pred = km.predict(pred_auto)
cm = confusion_matrix(pred, y_test)

#Reordering confusion matrix with correct labels 
indexes = linear_assignment(-cm + np.max(cm))
js = [e[1] for e in sorted(indexes, key=lambda x: x[0])]
cm2 = cm[:, js]

#Evaluating Accuracy
accuracy = np.trace(cm2) / np.sum(cm2)
print('Accuracy is:', accuracy * 100)
print('Confusion Matrix is:')
print(cm2)
```

```python
#####TASK 3
import numpy as np
import keras
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, confusion_matrix, normalized_mutual_info_score
from sklearn.mixture import GaussianMixture
from keras.models import Model, Sequential
from keras.layers import Input, Dense, Conv2D, MaxPooling2D, UpSampling2D, Flatten, Reshape
from keras import regularizers
from keras.datasets import fashion_mnist

#Loading the dataset
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

#Data normalization
max_value = float(x_train.max())
x_train = x_train.astype('float32') / max_value
x_test = x_test.astype('float32') / max_value

#Data preprocessing
x_train = x_train.reshape((len(x_train), np.prod(x_train.shape[1:])))
x_test = x_test.reshape((len(x_test), np.prod(x_test.shape[1:])))

#Building autoencoder
input_dim = x_train.shape[1]
encoding_dim = 32

autoencoder = Sequential()

#Encoder layers
autoencoder.add(Dense(4 * encoding_dim, input_shape=(input_dim,), activation='relu'))
autoencoder.add(Dense(2 * encoding_dim, activation='relu'))
autoencoder.add(Dense(encoding_dim, activation='relu'))

#Decoder layers
autoencoder.add(Dense(2 * encoding_dim, activation='relu'))
autoencoder.add(Dense(4 * encoding_dim, activation='relu'))
autoencoder.add(Dense(input_dim, activation='sigmoid'))

#Encoder Model
input_img = Input(shape=(input_dim,))
encoder_layer1 = autoencoder.layers[0]
encoder_layer2 = autoencoder.layers[1]
encoder_layer3 = autoencoder.layers[2]
encoder = Model(input_img, encoder_layer3(encoder_layer2(encoder_layer1(input_img))))

#Model Compilation
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')
history_train = autoencoder.fit(x_train, x_train,
                epochs=10,
                batch_size=256,
                validation_data=(x_test, x_test))

#Data visualization
loss = history_train.history['loss']
val_loss = history_train.history['val_loss']
epochs = range(10)
plt.figure()
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.ylabel('Loss') 
plt.xlabel('Epochs') 
plt.legend()
plt.show()

#Auto Encoder with GMM clustering
pred_auto_train = encoder.predict(x_train)
pred_auto = encoder.predict(x_test)
gmm = GaussianMixture(n_components=10, covariance_type='full', tol=0.001, random_state = 1)
gmm.fit(pred_auto_train)
pred = gmm.predict(pred_auto)
cm = confusion_matrix(pred, y_test)

#Reordering confusion matrix with correct labels
indexes = linear_assignment(-cm + np.max(cm))
js = [e[1] for e in sorted(indexes, key=lambda x: x[0])]
cm2 = cm[:, js]

#Evaluating Accuracy
accuracy = np.trace(cm2) / np.sum(cm2)
print('Accuracy is:', accuracy * 100)
print('Confusion Matrix is:')
print(cm2)
```
